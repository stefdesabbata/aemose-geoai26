```python
import pandas as pd
from tqdm import tqdm
```

```python
dataset_path        = 'data/processed/28714514_2021UKCensus_processed_20251009001.parquet'
shuffled_path       = 'data/processed/28714514_2021UKCensus_processed_20251009001rs.parquet'
```

```python
ds = pd.read_parquet(dataset_path)
rs = pd.read_parquet(shuffled_path)
```

```python
ds.head()
```

```python
rs.head()
```

```python
ds.loc["E00001096"]
```

```python
rs.loc["E00001096"]
```

```python
assert ds.loc["E00001096"].equals(rs.loc["E00001096"]), f"OA E00001096 does not match between ds and rs"
```

```python
oas = ds.index.to_list()
```

```python
for oa in tqdm(oas):
    ds_oa = ds.loc[oa]
    rs_oa = rs.loc[oa]
    assert ds_oa.equals(rs_oa), f"OA {oa} does not match between ds and rs"
```
