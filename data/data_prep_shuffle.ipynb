```python
import pandas as pd
import random
```

```python
dataset_path        = '../data/processed/28714514_2021UKCensus_processed_20251009001.parquet'
dataset_nickname    = 'ukc2021-proc20251009001'
shuffled_path       = '../data/processed/28714514_2021UKCensus_processed_20251009001rs.parquet'
```

```python
input_dataset = pd.read_parquet(dataset_path)
print(f"Dataset index: {input_dataset.index.name}, shape: {input_dataset.shape}")
input_dataset
```

```python
seed = random.randint(100, 100000)
seed
```
