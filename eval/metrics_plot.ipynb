```python
from tensorboard.backend.event_processing import event_accumulator
import matplotlib.pyplot as plt
```

```python
log_dir = 'scratch/ukc2021-proc20251009001rs__aemose-4-4_e7x3_s192-128-96-64_sage128_g0-64-32-g1-32-16_b512_20260517124211/logs/log_ukc2021-proc20251009001rs__aemose-4-4_e7x3_s192-128-96-64_sage128_g0-64-32-g1-32-16_b512_20260517124211/version_0'
```

```python
tb_ea = event_accumulator.EventAccumulator(log_dir)
tb_ea.Reload()
print("Available metrics:", tb_ea.Tags()['scalars'])
```

```python
metric = 'loss'

train_steps = [event.step for event in tb_ea.Scalars('train_' + metric)]
train_values = [event.value for event in tb_ea.Scalars('train_' + metric)]
val_steps = [event.step for event in tb_ea.Scalars('val_' + metric)]
val_values = [event.value for event in tb_ea.Scalars('val_' + metric)]

plt.figure(figsize=(10, 6))
plt.plot(train_steps, train_values, label='Training ' + metric)
plt.plot(val_steps, val_values, label='Validation ' + metric)
plt.xlabel('Steps')
plt.ylabel(metric)
plt.title(f'Training and Validation {metric}')
plt.legend()
plt.axvline(x=516868, color='red', linestyle='--', linewidth=1, alpha=0.7, label='Best Checkpoint')
plt.grid(True)
plt.ylim(0.0, 0.05)
plt.show()
```
